### Cleaning data

```python
#importing libraries
import csv
import pandas as pd
```

```python
#Importing all csv
oilCrawl = pd.read_csv("../../Milestone1/oilCrawl/oilCrawl.csv")
goldRate = pd.read_csv("../../Milestone1/goldCrawler/goldRate.csv")
currencyExchange = pd.read_csv("../../Milestone1/currencyCrawl/currencyExchange.csv")
```

```python
#Viewing oilCrawl
oilCrawl.head()
```

```python
#Removing the symbol '%' and '$'
oilCrawl['Annual% Change'] = list(map(lambda x: x[:-1], oilCrawl['Annual% Change'].values))
oilCrawl['AverageClosing Price'] = list(map(lambda x: x[1:], oilCrawl['AverageClosing Price'].values))
oilCrawl['Year Open'] = list(map(lambda x: x[1:], oilCrawl['Year Open'].values))
oilCrawl['Year High'] = list(map(lambda x: x[1:], oilCrawl['Year High'].values))
oilCrawl['Year Low'] = list(map(lambda x: x[1:], oilCrawl['Year Low'].values))
oilCrawl['Year Close'] = list(map(lambda x: x[1:], oilCrawl['Year Close'].values))

#We note that column values for columns have data types of string
#(even after the '%' and '$' symbol has been removed), so we need to convert these to numeric data as follows:
oilCrawl['Annual% Change'] = [float(x) for x in oilCrawl['Annual% Change'].values]
oilCrawl['AverageClosing Price'] = [float(x) for x in oilCrawl['AverageClosing Price'].values]
oilCrawl['Year Open'] = [float(x) for x in oilCrawl['Year Open'].values]
oilCrawl['Year High'] = [float(x) for x in oilCrawl['Year High'].values]
oilCrawl['Year Low'] = [float(x) for x in oilCrawl['Year Low'].values]
oilCrawl['Year Close'] = [float(x) for x in oilCrawl['Year Close'].values]

oilCrawl.head()
```

```python
#Cleaning oilCrawl
clean_index_oilCrawl = oilCrawl.drop(columns=['No'])
new_oilCrawl = clean_index_oilCrawl.rename(columns={"AverageClosing Price":"Price(USD)", "Year Open":"Open(USD)", 
                                                    "Year High":"High(USD)","Year Low":"Low(USD)", 
                                                    "Year Close":"Close(USD)", "Annual% Change":"Change %"})
new_oilCrawl.head()
```

```python
#Exporting new csv after data are cleaned
new_oilCrawl.to_csv('cleanOilRate.csv',mode = 'w', index=False)
```

```python
#Viewing goldRate
goldRate.head()
```

```python
#Removing the symbol '%' and '$'
goldRate['Change %'] = list(map(lambda x: x[:-1], goldRate['Change %'].values))
goldRate['Vol.'] = list(map(lambda x: x[:-1], goldRate['Vol.'].values))

#We note that column values for columns have data types of string
#(even after the '%' and '$' symbol has been removed), so we need to convert these to numeric data as follows:
goldRate['Change %'] = [float(x) for x in goldRate['Change %'].values]
goldRate['Vol.'] = [float(x) for x in goldRate['Vol.'].values]

# Converting K digits to thousand decimal
goldRate['Vol.'] = goldRate['Vol.'].mul(1000)

goldRate.head()
```

```python
#Cleaning goldRate
new_goldRate = goldRate.rename(columns={"Date":"MonthYear","Price":"Price(USD)", "Open":"Open(USD)", 
                                                    "High":"High(USD)","Low":"Low(USD)", 
                                                    "Vol.":"Volume(USD)", "Change %":"Change %"})
new_goldRate.head()
```

```python
#Exporting new csv after data are cleaned
new_goldRate.to_csv('cleanGoldRate.csv',mode = 'w', index=False)
```
