```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
#from mlxtend.frequent_patterns import apriori
#from mlxtend.frequent_patterns import association_rules
```

```python
geroceries = [['Milk', 'Onions','Butter','Oats','Eggs', 'Yogurt'],
             ['Spinach', 'Onions','Butter','Oats','Eggs', 'Yogurt'],
              ['Milk', 'Bread','Corn','Oats','Yogurt'],
              ['Milk', 'Apple','Oats','Eggs'],
              ['Corn', 'Onions','Oats','Ice cream', 'Eggs'],
              ['Milk', 'Apple','Oats','Eggs'],
              ['Spinach', 'Onions','Butter','Oats','Eggs', 'Yogurt'],
              ['Milk', 'Bread','Corn','Oats','Yogurt'],
              ['Corn', 'Onions','Bananas','Oats','Ice cream', 'Eggs']]
```

```python
item_dict = {}
for items in geroceries:
    for item in items:
        if item not in item_dict:
            item_dict[item]=0
            
        item_dict[item]+= 1
```

```python
item_dict
```

```python
item_names = []
item_frequencies = []
for key, val in item_dict.items():
    item_names.append(key)
    item_frequencies.append(val)
```

```python
items_table = pd.DataFrame({'Names': item_names,
                           'Frequencies': item_frequencies})
items_table
```

```python
items_table.sort_values('Frequencies', ascending=False)
```

```python
items_table.plot.bar(y='Frequencies', x='Names', figsize=(12,8))
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules

transencoder = TransactionEncoder()
transencoder_array = transencoder.fit(geroceries).transform(geroceries)
```

```python
geroceries_df = pd.DataFrame(transencoder_array, columns=transencoder.columns_)
geroceries_df
```

```python
geroceries_df.shape
```

```python
num_records = geroceries_df.shape[0]
num_records
```
