```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import numpy as np
import csv
from pathlib import Path
```

```python
csvExist = Path('goldRate.csv') #creating a file in csv extention
#Only enable when overwriting with new columns
#if csvExist.is_file():
#    pass
#else:
with open ('goldRate.csv','wb') as createFile: #This line will create a csv file under name webcrawler
        filewriter = csv.writer(createFile)

url = "https://www.investing.com/commodities/gold-historical-data"
req = urllib.request.Request(url, data=None, headers={'User-Agent': 'Chrome/35.0.1916.47'})

soup = BeautifulSoup(urllib.request.urlopen(req).read(),"lxml")

#extract data 
rows = soup.find('table',{'class': 'genTbl closedTbl historicalTbl'}).findAll('tr')[1:]
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip(' ') for ele in cols]
    data.append([ele for ele in cols if ele])
 
#extract column names 
colnames = soup.find('table',{'class': 'genTbl closedTbl historicalTbl'}).findAll('tr')[:1]
col_names = []
for col in colnames:
    cols = col.find_all('th')
    cols = [ele.text.strip() for ele in cols]
    col_names.append(cols)
col_names = col_names[0]
    
#Write data to files
df1 = pd.DataFrame(data,columns = col_names)
df1['Change %'] = list(map(lambda x: x[:-1], df1['Change %'].values))
df1['Vol.'] = list(map(lambda x: x[:-1], df1['Vol.'].values))
df1=df1.rename(columns={"Date":"Date","Price":"PriceUSD", "Open":"OpenUSD", 
                                                    "High":"HighUSD","Low":"LowUSD", 
                                                    "Vol.":"VolumeUSD", "Change %":"Change"})
df1.to_csv('goldRate.csv',mode = 'w', index=False)
df1.head()
```

```python
# importing module 
import sqlite3 
import csv
  
# connecting to the database  
connection = sqlite3.connect("wqd7005_final.db") 
  
# cursor  
crsr = connection.cursor() 

# SQL command to drop a table in the database if exist
crsr.execute("DROP TABLE IF EXISTS goldRate;")
crsr.execute("DROP TABLE IF EXISTS currencyExchangeMYR2USD;")
crsr.execute("DROP TABLE IF EXISTS borneoOilUSD;")


# SQL command to create a table in the database 
createTableA = """CREATE TABLE goldRate (  
Date DATETIME,
PriceUSD FLOAT,
OpenUSD FLOAT,
HighUSD FLOAT,
LowUSD FLOAT,
VolumeUSD FLOAT,
Change FLOAT);"""

createTableB = """CREATE TABLE currencyExchangeMYR2USD (  
Date DATETIME,
PriceUSD FLOAT,
OpenUSD INTEGER,
HighUSD FLOAT,
LowUSD INTEGER,
Change FLOAT);"""

createTableC = """CREATE TABLE borneoOilUSD (  
Date DATETIME,
PriceUSD FLOAT,
OpenUSD FLOAT,
HighUSD FLOAT,
LowUSD FLOAT,
VolumeUSD FLOAT,
Change FLOAT);"""
  
# execute the statement createTable
crsr.execute(createTableA)
crsr.execute(createTableB)
crsr.execute(createTableC)

# SQL command to list down the table details 
crsr.execute("PRAGMA table_info(goldRate);")
crsr.execute("PRAGMA table_info(currencyExchangeMYR2USD);")
crsr.execute("PRAGMA table_info(borneoOilUSD);")

read_data = pd.read_csv((r'goldRate.csv'), sep=",",index_col=[0])
read_data = pd.read_csv((r'currencyExchangeMYR2USD.csv'), sep=",",index_col=[0])
read_data = pd.read_csv((r'borneoOilUSD.csv'), sep=",",index_col=[0])
 # Insert the values from the csv file into the table
read_data.to_sql('goldRate', connection, if_exists='append', index = False)
read_data.to_sql('currencyExchangeMYR2USD', connection, if_exists='append', index = False)
read_data.to_sql('borneoOilUSD', connection, if_exists='append', index = False)

# execute the command to fetch all the data from the table emp 
crsr.execute("SELECT * FROM goldRate LIMIT 3;")  

# store all the fetched data in the ans variable 
ans = crsr.fetchall() 

print("SELECT * FROM goldRate LIMIT 5;  \n",ans)


# To save the changes in the files. Never skip this.  
# If we skip this, nothing will be saved in the database. 
connection.commit() 
  
# close the connection 
connection.close() 
```
